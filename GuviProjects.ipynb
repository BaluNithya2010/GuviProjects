```python
import pandas as pd
```

```python
csv_file_path = "orders.csv"
df = pd.read_csv(csv_file_path)
df.head()
```

```python
df_renamed = df.rename(columns = {
    "Order Id" : "order_id",
    "Order Date" : "order_date",
    "Ship Mode" : "ship_mode",
    "Segment" : "segment",
    "Country" : "country",
    "City" : "city",
    "State" : "state",
    "Postal Code" : "postal_code",
    "Region" : "region",
    "Category" : "category",
    "Sub Category" : "sub_category",
    "Product Id" : "product_id",
    "cost price" : "cost_price",
    "List Price" : "list_price",
    "Quantity" : "quantity",
    "Discount Percent" : "discount_percent"
})
```

```python
df_renamed
```

```python
#removing trailing space from text columns
df_renamed = df_renamed.apply(lambda x: x.str.rstrip() if x.dtype == 'Object' else x)
```

```python
df_renamed['discount'] = df_renamed['list_price'] * (df_renamed['discount_percent'] / 100)
```

```python
df_renamed['sale_price'] = df_renamed['list_price'] - df_renamed['discount']
```

```python
df_renamed['profit'] = (df_renamed['sale_price'] - df_renamed['cost_price']) * df_renamed['quantity']
```

```python
df_renamed
```

```python
df_orders = df_renamed[['order_id','order_date','ship_mode','segment','country','city','state','postal_code','region']]
```

```python
df_orders
```

```python
df_order_details = df_renamed[['order_id','category','sub_category','product_id','cost_price','list_price','quantity','discount_percent','discount','sale_price','profit']]
```

```python
import pymysql

# MySQL connection details
username = 'root'  # Replace with your MySQL username
password = 'T1e2s3t4'  # Replace with your MySQL password
database = 'my_database'  # Replace with your MySQL database name
host = 'localhost'          # MySQL server host
port = 3306                 # MySQL server port (default: 3306)

# Connect to MySQL
connection = pymysql.connect(
    host=host,
    user=username,
    password=password,
    database=database,
    port=port
)

# Create cursor object to execute SQL queries
cursor = connection.cursor()

# Create 'orders' table with primary key
create_orders_table = """
CREATE TABLE IF NOT EXISTS orders (
    order_id INT PRIMARY KEY,
    order_date DATE NOT NULL,
    ship_mode VARCHAR(50),
    segment VARCHAR(50),
    country VARCHAR(50),
    city VARCHAR(50),
    state VARCHAR(50),
    postal_code INT,
    region VARCHAR(50)
);
"""

# Create 'products' table with foreign key referencing 'orders' table
create_order_details_table = """
CREATE TABLE IF NOT EXISTS order_details (
    product_id VARCHAR(50),
    order_id INT NOT NULL,
    category VARCHAR(50),
    sub_category VARCHAR(50),
    cost_price DECIMAL(10, 2),
    list_price DECIMAL(10, 2),
    quantity INT,
    discount_percent INT,
    discount INT,
    sale_price DECIMAL(10, 2),
    profit INT,
    FOREIGN KEY (order_id) REFERENCES orders(order_id) 
        ON DELETE CASCADE
        ON UPDATE CASCADE
);
"""

try:
    # Execute the SQL statements to create tables
    cursor.execute(create_orders_table)
    cursor.execute(create_order_details_table)
    
    # Commit changes
    connection.commit()

    print("Tables created successfully with primary and foreign key constraints.")
except Exception as e:
    print(f"Error: {e}")
    connection.rollback()
finally:
    # Close the connection
    cursor.close()
    connection.close()
```

```python
import mysql.connector
from mysql.connector import Error

def create_database():
    try:
        # Connect to the MySQL server (not specifying a database yet)
        connection = mysql.connector.connect(
            host='localhost',  # or your MySQL server IP
            user='root',       # your MySQL username
            password='T1e2s3t4'  # your MySQL password
        )
        
        if connection.is_connected():
            # Create a cursor object to interact with MySQL
            cursor = connection.cursor()
            
            # Create the new database
            cursor.execute("CREATE DATABASE IF NOT EXISTS my_database")
            print("Database 'my_database' created successfully.")
            
    except Error as e:
        print(f"Error while creating database: {e}")
    
    finally:
        # Close the connection
        if connection.is_connected():
            cursor.close()
            connection.close()

# Call the function to create the database
create_database()
```
